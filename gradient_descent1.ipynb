## Gradient Descent Implementation #1
Here we implement the most basic code of estimating the parameters of a line $y = mx + b$.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

To test our approach, we assign arbitrary values to the parameters we wish to estimate and generate data with them. Noise is added to the data to simulate data that may have been collected via some procedure.

```python
x = np.linspace(-100, 100, 100)
n = len(x)

#arbitrary parameter values
m = 4.
b = -2.

#generated data
y = m*x + b + np.random.normal(0,50,n)

plt.scatter(x,y)
plt.show()
```

At each iteration, the error is the difference between the predicted values (using current parameter estimates) and the actual values.  The error is used to compute a cost and the parameter updates.

The error is $y - y_p = y - \theta_1x - \theta_0$.

Since our goal is to minimize the (squared) error with respect to the parameter estimates, application of the gradient descent method yields the following parameter updates:

$\theta_0 = \theta_0 - learning rate * error$

$\theta_1 = \theta_1 - learning rate * error * x$


```python
learning_rate = 1e-7
iterations = 300

def cost(error):
    N = len(error)
    return 1./N * error.dot(error)

theta1 = np.random.uniform(0., 4.)
theta0 = np.random.uniform(-3., 3.) 
costs = []

for _ in range(iterations):
    yp = theta1*x + theta0
    error = yp - y
    costs.append(cost(error))
    theta1 -= learning_rate * error.dot(x)
    theta0 -= learning_rate * error.sum()


print(costs[-1])
print(m, b)
print(theta1, theta0)

#plot costs
plt.plot(costs)
plt.show()

#plot the data with the parameter estimates
plt.scatter(x,y)
plt.plot(x, theta1*x + theta0, linewidth=3, color='red')
plt.show()
```
